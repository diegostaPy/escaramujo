```python
import ARLreader as Ar
#import matplotlib.pyplot as plt
import datetime
import numpy as np
import pandas as pd
```

#ftp://arlftp.arlhq.noaa.gov/pub/archives/gdas1

```python
dt = datetime.datetime(2021, 4, 6)
end = datetime.datetime(2021, 4, 20)
step = datetime.timedelta(hours=3)
dates= []
while dt <= end:
    dates.append(dt)
    dt += step
```

```python
print()
```

current7days es la primera semana de abril

```python
Teff=[]
Tg=[]
Pg=[]
for date in dates:
    print(date)
    filename=Ar.fname_from_date(date)
    if(date.day>=8 and date.day<=14 ):
        filename='gdas1.apr21.w2'
    if(date.day>=15 and date.day<=21):
        filename='gdas1.apr21.w3'
    print(filename)
    print(date.day)
    profile, sfcdata, indexinfo, ind = Ar.reader('gdas/'+filename).load_profile(date.day, date.hour, (16.75973,-93.11308),sfc=True, flag_interp=True)
#    profile, sfcdata, indexinfo, ind = Ar.reader('gdas/'+filename).load_profile(date.day, date.hour, ( -25.331270,-57.516597),sfc=True, flag_interp=True)
  
   
    T=profile['TEMP']#- 273.15
    P=profile['PRSS']    
    H=profile['HGTS']
    Ts=sfcdata['TMPS']    
    Ps=sfcdata['PRSS']

    Wun=-np.diff(np.hstack((P,0)))/P[0]
    Teff.append(np.sum(Wun*T))
    Tg.append(Ts)
    Pg.append(Ps)
    
```

```python
dataTeff = pd.DataFrame(list(zip( dates, Teff,Tg,Pg)), 
               columns =['date', 'teff', 'ts', 'ps']) 
```

```python
dataTeff.to_csv("teffDatavAbrMex2021.csv")
```
